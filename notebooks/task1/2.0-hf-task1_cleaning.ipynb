# Imports

```python
import os
import numpy as np
import pandas as pd
```

```python
from data_gadgets.cleaning import Cleaner
```

# Reading Data

```python
path = os.path.join('..', '..', 'data', 'raw', 'data_task1.csv')
data = pd.read_csv(path)
```

```python
data.head()
```

# Cleaning Names

```python
cleaner = Cleaner()
```

```python
data = cleaner.headers(data)
```

```python
cols = cleaner.separate_data(data, 'survived')
```

```python
data = cleaner.categories(data)
```

```python
data.head()
```

# Filling Missing Values

```python
data['age'] = data.age.fillna(data.age.median())
data['fare'] = data.fare.fillna(data.fare.median())
```

```python
cabin = []
for i in data.cabin:
    cabin.append(str(i)[0].title())
data['cabin'] = cabin
```

```python
data.head()
```

# Merging Values

```python
sibsp = []
for value in data.sibsp:
    if value > 1:
        value = 1
    sibsp.append(value)
data['sibsp'] = sibsp
```

```python
parch = []
for value in data.parch:
    if value > 1:
        value = 1
    parch.append(value)
data['parch'] = parch 
```

```python
family = []
for value1, value2 in zip(data.sibsp, data.parch):
    if value1 == 1 or value2 == 1:
        family.append(1)
    else:
        family.append(0)
data['family'] = family
```

```python
data.head()
```

# Dropping Values

```python
data = data.drop(['name', 'passengerid', 'ticket', 'sibsp', 'parch'], axis=1)
```

```python
data.head()
```

# Encoding Categories

```python
data = pd.get_dummies(data, drop_first=True, dtype=int)
```
